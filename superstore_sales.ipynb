### Time-Based Performance Analysis

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d/%m/%Y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%d/%m/%Y')
```

```python
data['Year'] = data['Order Date'].dt.year
```

```python
data['Order Date'].head()
```

```python
yearly_sales = data.groupby('Year')['Sales'].sum()
```

```python
yearly_sales.plot(kind='line', marker='o')
plt.title("Sales by Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
data['Month']=data['Order Date'].dt.month
```

```python
monthly_sales=data.groupby('Month')['Sales'].sum()
```

```python
monthly_sales.plot(kind="line",marker='o')
plt.title("Sales by Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
yearly_sales = data.groupby('Year')['Sales'].sum()
growth = yearly_sales.pct_change() * 100
print(growth)
```

### Location-Based Sales Analysis

```python
state_sales = data.groupby('State')['Sales'].sum().sort_values(ascending=False)

state_sales.head()
```

```python
state_sales.head(10).plot(kind='bar')
plt.title("Top 10 States by Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
region_sales = data.groupby('Region')['Sales'].sum()
region_sales.plot(kind='bar')
plt.title("Region Performance")
plt.ylabel("Total Sales")
plt.show()
```

### Customer Analysis

```python
top_customers = data.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)
top_customers.head(10)
```
